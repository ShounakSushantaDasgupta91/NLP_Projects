```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# 1. Download ALL necessary resources
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

text = "Google is headquartered in Mountain View, California."

# 2. Pipeline: Tokenize -> POS Tag -> NE Chunk
words = word_tokenize(text)
tag_elements = pos_tag(words)
chunked_tree = ne_chunk(tag_elements)

# 3. Draw (Requires Tkinter installed on your OS)
try:
    chunked_tree.draw()
except Exception as e:
    print(f"Drawing failed. You might be missing Tkinter. Error: {e}")
    # Fallback: Print the tree structure in text format
    print(chunked_tree)
```
